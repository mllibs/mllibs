```python
#!/usr/bin/env python3

import requests
from bs4 import BeautifulSoup

# review_link = 'https://www.edlitera.com/en/blog/posts/pyspark-cheat-sheet'
review_link = "https://sparkbyexamples.com/pyspark/pyspark-withcolumn/"
r = requests.get(review_link)
page = r.text
```

```python
page
```

```python
soup = BeautifulSoup(page, 'html.parser')

# review summary
content = soup.find_all(class_='language-python')

for i in content:
    print(i.text,'\nend')
```

```python
user_id = soup.find(text='#mcetoc_')
print(user_id)
```

```python
all_string=soup.find_all("h2")

lst_data = []
for i in all_string:
    if("\xa0" not in i.text):
        lst_data.append(i)

lst_data
```
